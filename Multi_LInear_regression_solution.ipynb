```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
# Load the data
X = []
Y = []

for line in open('multi_linear_regression_data.csv'):
  x1, x2, y = line.split(',')
  # remember we are adding the 1 as a bias term of x0
  X.append([float(x1), float(x2), 1])
  Y.append(float(y))
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
# Lets plot the data first

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
plt.show()
```

```python
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T,Y))
Yhat = np.dot(X,w)
```

```python
# compute r-squared
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - (d1.dot(d1) / d2.dot(d2))
print('r-squared: ', r2)
```
